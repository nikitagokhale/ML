```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def objectiveFunction(x):
    return (x+3)**2
```

```python
def derivative(x):
    return 2*(x+3)
```

```python
def gradDescent(x, rate, n):
    xValues = []
    yValues = []
    
    for _ in range(n):
        x = x - rate*derivative(x)
        xValues.append(x)
        yValues.append(objectiveFunction(x))
        
    return xValues, yValues
```

```python
initialX = 2
learningRate = 0.1
noIterations = 50
```

```python
xValues, yValues = gradDescent(initialX, learningRate, noIterations)
```

```python
finalX = xValues[-1]
finalY = yValues[-1]
print(finalX)
print(finalY)
```

```python
xRange = np.linspace(-8,2,100)
yRange = objectiveFunction(xRange)
```

```python
plt.plot(xRange, yRange, label='Function: $y=(x+3)^2$')
plt.scatter(xValues, yValues, color='red', label='Optimization Path')
plt.title('Gradient Descent Optimization')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
